```python
import numpy as np
from scipy import stats

numerik = df.select_dtypes(include=['number']).columns.tolist()
kategori = df.select_dtypes(exclude=['number']).columns.tolist()
```

```python
Q1 = df[numerik ].quantile(0.25)
Q3 = df[numerik ].quantile(0.75)
IQR= Q3-Q1
#tentukan batas bawah dan atas

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_outlier = df[((df[numerik]< lower_bound) | (df[numerik]>upper_bound)).any(axis=1)]

print('data outlier :', df_outlier.shape[0],"baris")
```

```python
df_no_outlier = df[~((df[numerik]< lower_bound) | (df[numerik]>upper_bound)).any(axis=1)]

#kedepannya memakai df_no_outlier
print ('data sebelum :',df.shape[0],'baris')
print ('data sesudah :',df_no_outlier.shape[0],'baris')
```

```python
def age_bucket(age):
    if age < 25: return "young"
    elif age < 35: return "adult"
    else: return "mature"

df['age_bucket'] = df['age'].apply(age_bucket)
df
```

```python
df['device_browser'] = df['device'] + "_" + df['browser']

df['is_mobile'] = df['device'].isin(['Mobile']).astype(int)
df
```

```python
rata_usia_session_duration = df_no_outlier.groupby('age')['session_duration'].mean().reset_index()
rata_usia_session_duration
```

```python
deskripsi_gender = df_no_outlier.groupby('gender')['page_view'].agg(['mean','min','max','count',]).reset_index()
deskripsi_gender
```
